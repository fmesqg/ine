## Pobreza

https://www.ine.pt/xportal/xmain?xpid=INE&xpgid=ine_indicadores&indOcorrCod=0009821&contexto=bd&selTab=tab2

```python
from ine_api import get_df, default_plot

anos = ",".join([f"S7A{ano}" for ano in range(2018, 2024)])
params_base = {
    "lang": "PT",
    "Dim1": anos,
}
params_pobreza = {**params_base, "varcd": "0009821"}

df_pobreza = get_df(params_pobreza)
default_plot(
    df=df_pobreza, title="Taxa de risco de pobreza (Após transferências sociais - %)"
)
```
